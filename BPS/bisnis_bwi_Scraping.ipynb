```python
import pandas as pd
import time
import random
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# --- CONFIG TEST ---
KEYWORD = "Cafe Banyuwangi"
SCROLL_COUNT = 10 

chrome_options = Options()
# chrome_options.add_argument("--headless") # Matikan comment ini jika ingin browser tidak terlihat
chrome_options.add_argument("--disable-notifications")
prefs = {"profile.managed_default_content_settings.images": 2} # Matikan gambar biar ngebut
chrome_options.add_experimental_option("prefs", prefs)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

def smart_sleep(min_sec, max_sec):
    time.sleep(random.uniform(min_sec, max_sec))

print(f"üöÄ Memulai Test Drive untuk: {KEYWORD}")
start_time = time.time() # Mulai itung waktu

try:
    driver.get(f"https://www.google.com/maps/search/{KEYWORD}")
    smart_sleep(5, 7)

    # 1. Proses Scrolling
    print("üñ±Ô∏è Sedang men-scroll daftar...")
    for i in range(SCROLL_COUNT):
        try:
            # Selektor panel kiri Maps
            panel = driver.find_element(By.XPATH, '//div[@role="feed"]')
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', panel)
            print(f"   [{i+1}/{SCROLL_COUNT}] Scrolling...")
            smart_sleep(2, 4)
        except:
            print("Gagal scroll, mungkin data sudah mentok.")
            break

    # 2. Ekstraksi Data
    print("üì¶ Mengambil data nama bisnis...")
    elements = driver.find_elements(By.CLASS_NAME, "hfpxzc")
    
    all_results = []
    for el in elements:
        name = el.get_attribute("aria-label")
        alamat = el.get_dom_attribute("span")
        link = el.get_attribute("href")
        if name:
            all_results.append({"nama_bisnis": name, "alamat": alamat, "link": link})

    # 3. Simpan ke Excel
    df = pd.DataFrame(all_results)
    df.to_excel("test_cafe_bwi.xlsx", index=False)

    end_time = time.time()
    durasi = end_time - start_time
    
    print("-" * 30)
    print(f"‚úÖ BERHASIL!")
    print(f"üìä Total Data: {len(df)}")
    print(f"‚è±Ô∏è Waktu Tempuh: {round(durasi, 2)} detik")
    print(f"üìÇ File: test_cafe_bwi.xlsx")
    print("-" * 30)

finally:
    driver.quit()
```
