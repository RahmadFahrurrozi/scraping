```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=chrome_options)

data_berita = []

try:
    print("Starting web scraping open news detik.com...")
    driver.get("https://www.detik.com/terpopuler")
    time.sleep(3)
    
    articles = driver.find_elements(By.TAG_NAME, "article")
    print(f"Found {len(articles)} articles. Extracting data...")
    
    for i, articles in enumerate(articles[:20]):
        try:
            title_element = articles.find_element(By.CLASS_NAME, "media__title")
            title = title_element.text
            
            link_elment = title_element.find_element(By.TAG_NAME, "a")
            link = link_elment.get_attribute("href")
            
            data_berita.append({
                "No": i + 1,
                "Title": title,
                "Link": link
            })
            print(f"Extracted article {i + 1}: {title}")
        except Exception as e:
            continue
        
        if data_berita:
            df = pd.DataFrame(data_berita)
            file_name = "detik_terpopuler.xlsx"
            df.to_excel(file_name, index=False)
            print(f"Data successfully saved to {file_name}")
        else:
            print("No articles found. nothing to save.")
    
finally:
    time.sleep(2)
    driver.quit()
```
